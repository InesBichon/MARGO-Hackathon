```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("train.csv")
```

```python
len(df.columns)
```

```python
ranges = (df.columns[1:200], df.columns[200:2248], df.columns[2248:4296])
```

```python
df.iloc[:,3:5]
```

```python
def knn(k, r):
	X = df[r][:,:-1].values
	y = df.iloc[:,-1].values

	X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

	knn_model = KNeighborsClassifier(n_neighbors=k)
	knn_model.fit(X_train, y_train)

	y_pred = knn_model.predict(X_test)

	acc = accuracy_score(y_test, y_pred)
	print (f"Précision: {acc:.2f}")
	print ("Rapport de classification:")
	print (classification_report(y_test, y_pred))

knn(5, ranges[1])
```

```python
def kappa_cohen(a, b, c, d):
	tot = a + b + c + d
	p_acc = (a + d) / tot
	p_has = ((a + b) * (a + c) + (c + d) * (b + d)) / tot**2
	return (p_acc - p_has) / (1 - p_has)
```

```python
from sklearn.feature_selection import VarianceThreshold

X_base = df.iloc[:, 1:-1]
y = df.iloc[:, -1]
```

```python
X_base.head()
```

```python
y.head()
```

```python
cols = X_base.columns
X = X_base.copy()
for i in range(199):
	X.iloc[:,i] = (X.iloc[:,i] - X.iloc[:,i].mean()) / X.iloc[:,i].std()

X.iloc[:, 199:2247] = (X.iloc[:, 199:2247] - X.iloc[:, 199:2247].mean()) / X.iloc[:, 199:2247].std()
X.iloc[:, 2247:] = (X.iloc[:, 2247:] - X.iloc[:, 2247:].mean()) / X.iloc[:, 2247:].std()

X.head()
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.01)  # Supprime les features avec une variance < 0.01
reduced_X = selector.fit_transform(X)
reduced_Xb = selector.fit_transform(X_base)
```

```python
print (reduced_X.shape)
print (X.shape)
print (reduced_Xb.shape)
```

```python
from sklearn.decomposition import PCA

n_col = 200

pca = PCA(n_components=n_col)
reduced_data = pca.fit_transform(reduced_X)
```

```python
reduced_df = pd.DataFrame(reduced_data, columns = [f'Component_{i+1}' for i in range(n_col)])
reduced_df['class'] = y
```
